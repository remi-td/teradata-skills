## 2. Pick a table to explore

Set `${schema}` and `${table}` to the object you want to investigate.

```
%var schema=<schema>, table=<table>
```

### Row count & sample

Start by getting a feel for size and shape.

```
SELECT COUNT(*) AS row_count FROM ${schema}.${table};
```

```
SELECT TOP 10 * FROM ${schema}.${table};
```

### Column types

Pull the column metadata from DBC.

```
SELECT ColumnName, ColumnType, ColumnLength, Nullable
FROM DBC.ColumnsV
WHERE DatabaseName = '${schema}' AND TableName = '${table}'
ORDER BY ColumnId;
```

### Null & distinct profile

Edit the column list to match columns of interest.

```
SELECT
  COUNT(*)                                AS rows_total,
  COUNT(DISTINCT <col1>)                  AS distinct_col1,
  SUM(CASE WHEN <col1> IS NULL THEN 1 ELSE 0 END) AS nulls_col1
FROM ${schema}.${table};
```

### Distribution

Replace `<dimension>` and `<measure>` with appropriate columns, then visualize with `%chart`.

```
SELECT <dimension>, SUM(<measure>) AS total_measure
FROM ${schema}.${table}
GROUP BY <dimension>
ORDER BY total_measure DESC;
```

```
%chart <dimension>, total_measure, title=Distribution by dimension
```
