## 3. Setup: create a small teaching dataset

We use a tiny table so you can focus on the concept, not the data.

```
DROP TABLE teach_demo;
```

```
CREATE MULTISET TABLE teach_demo (
  id     INTEGER NOT NULL,
  region VARCHAR(20),
  amount DECIMAL(10,2)
) NO PRIMARY INDEX;
```

```
INSERT INTO teach_demo VALUES (1, 'North',  120.00);
INSERT INTO teach_demo VALUES (2, 'North',   80.00);
INSERT INTO teach_demo VALUES (3, 'South',  200.00);
INSERT INTO teach_demo VALUES (4, 'South',   50.00);
INSERT INTO teach_demo VALUES (5, 'East',   300.00);
```

## 4. The concept, step by step

### Step 1 — <first idea>

### Step 2 — <second idea>
