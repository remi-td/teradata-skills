## 3. Setup

Create the minimal objects needed for the demo.

```
DROP TABLE demo_sales;
```

```
CREATE MULTISET TABLE demo_sales (
  store_id INTEGER,
  sale_dt  DATE,
  amount   DECIMAL(10,2)
) NO PRIMARY INDEX;
```

```
INSERT INTO demo_sales VALUES (1, DATE '2025-01-01', 120.00);
INSERT INTO demo_sales VALUES (1, DATE '2025-01-02',  80.00);
INSERT INTO demo_sales VALUES (2, DATE '2025-01-01', 200.00);
INSERT INTO demo_sales VALUES (2, DATE '2025-01-02', 150.00);
INSERT INTO demo_sales VALUES (3, DATE '2025-01-01',  90.00);
```

## 4. The 'wow' moment

Show the single most compelling query. Keep it short — this is the screenshot-worthy cell.

```
SELECT store_id, SUM(amount) AS total_sales
FROM demo_sales
GROUP BY store_id
ORDER BY total_sales DESC;
```

```
%chart store_id, total_sales, title=Total Sales by Store, mark=bar
```

## 5. Variation — different angle

Same data, different lens — useful for answering 'what about X?' from the audience.

```
SELECT sale_dt, SUM(amount) AS daily_sales
FROM demo_sales
GROUP BY sale_dt
ORDER BY sale_dt;
```

```
%chart sale_dt, daily_sales, title=Sales by Day, mark=line, typex=t
```
